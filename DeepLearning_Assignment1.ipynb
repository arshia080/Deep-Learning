```python
import numpy as np
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
def train_perceptron(X, y, lr=0.1, epochs=20):
    w = np.zeros(X.shape[1])
    b = 0

    for _ in range(epochs):
        for i in range(len(X)):
            z = np.dot(w, X[i]) + b
            y_hat = step(z)
            error = y[i] - y_hat

            w = w + lr * error * X[i]
            b = b + lr * error

    return w, b
```

```python
def test_perceptron(X, w, b):
    outputs = []
    for x in X:
        z = np.dot(w, x) + b
        outputs.append(step(z))
    return outputs
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

AND  = np.array([0, 0, 0, 1])
OR   = np.array([0, 1, 1, 1])
NAND = np.array([1, 1, 1, 0])
NOR  = np.array([1, 0, 0, 0])
```

```python
w, b = train_perceptron(X, AND)
print("Weights:", w)
print("Bias:", b)
print("Predictions:", test_perceptron(X, w, b))
```

```python
w, b = train_perceptron(X, OR)
print(test_perceptron(X, w, b))

w, b = train_perceptron(X, NAND)
print(test_perceptron(X, w, b))

w, b = train_perceptron(X, NOR)
print(test_perceptron(X, w, b))
```
